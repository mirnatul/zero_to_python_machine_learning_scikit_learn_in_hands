Let's import some libraries (numpy and panda)

```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

Let's store the feature matrix in X and response vector in y

```python
X = iris.data
y = iris.target
```

Let's import the classifier (choose the model/estimator). Ex: KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

Let's focus on neighbor = 1

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

Let's train the model ðŸ‘Œ (fit the estimate)

```python
knn.fit(X, y)
```

Let's predict our first data (make prediction)

```python
knn.predict([[3, 5, 4, 2]])
```

```python
print("Predicted target name:", iris['target_names'][knn.predict([[3, 5, 4, 2]])])
```

Or, we can do this way,

```python
X_new = np.array([[3, 5, 4, 2]])
print("X_new.shape:", X_new.shape)
mypredict = knn.predict(X_new)
print("Prediction:", mypredict)
print("Predicted target name:", iris['target_names'][mypredict])
```

If we want to test two sample (we can just store the arrays in a variable instead for larger case)

```python
knn.predict([[3, 5, 4, 2], [5, 4, 3, 2]])
```

Let's evaluate when we consider neighbor = 3

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

Let's fit the model for 3

```python
knn.fit(X, y)
```

```python
knn.predict([[3, 5, 4, 2], [5, 4, 3, 2]])
```

Note:
When neighbors = 1, it gives (2, 1)
When neighbors = 3, it gives (1, 1)

Let's try Logistic Regression (learn more...)

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=110)
lr.fit(X, y)
```

```python
lr.predict([[5, 5, 4, 2], [5, 4, 3, 2]])
```
